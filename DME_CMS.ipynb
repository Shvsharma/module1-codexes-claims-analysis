```python
 import pandas as pd
```

```python
 data = pd.read_csv('https://data.cms.gov/sites/default/files/2023-04/67157de9-d962-4af0-bf0e-3578b3afec58/inpatient.csv' , sep='|')
```

```python
print(data.head())
```

```python
for column in data. columns: print(column)
```

```python
icd_prcdr_cd25 = data['ICD_PRCDR_CD25']
hcpcs_cd= data['HCPCS_CD']
icd_dgns_cd1 = data['ICD_DGNS_CD1']
icd_dgns_cd2 = data['ICD_DGNS_CD2']
icd_dgns_cd10 = data['ICD_DGNS_CD10']
icd_dgns_cd7 = data['ICD_DGNS_CD7']
icd_dgns_cd9 = data['ICD_DGNS_CD9']
icd_dgns_cd11 = data['ICD_DGNS_CD11']

```

```python
icd_frequency_cd25 = icd_prcdr_cd25.value_counts()
print(icd_frequency_cd25)
hcpcs_frequency_cd = hcpcs_cd.value_counts()
print(hcpcs_frequency_cd)
icd_frequency_cd1 = icd_dgns_cd1.value_counts()
print(icd_frequency_cd1)
icd_frequency_cd2 = icd_dgns_cd2.value_counts()
print(icd_frequency_cd2)
icd_frequency_cd10 = icd_dgns_cd10.value_counts()
print(icd_frequency_cd10)
icd_frequency_cd7 = icd_dgns_cd7.value_counts()
print(icd_frequency_cd7)
icd_frequency_cd9 = icd_dgns_cd9.value_counts()
print(icd_frequency_cd9)
icd_frequency_cd11 = icd_dgns_cd11.value_counts()
print(icd_frequency_cd11)
```

```python
missing_icd_cd25 = icd_prcdr_cd25.isnull().sum()
print(f"Missing ICD25 Codes: {missing_icd_cd25}")
missing_hcpcs_cd = hcpcs_cd.isnull().sum()
print(f"Missing HCPCS_CD Codes: {missing_hcpcs_cd}")
missing_icd_cd1 = icd_dgns_cd1.isnull().sum()
print(f"Missing ICD1 Codes: {missing_icd_cd1}")
missing_icd_cd2 = icd_dgns_cd2.isnull().sum()
print(f"Missing ICD2 Codes: {missing_icd_cd2}")
missing_icd_cd10 = icd_dgns_cd10.isnull().sum()
print(f"Missing ICD10 Codes: {missing_icd_cd10}")
missing_icd_cd7 = icd_dgns_cd7.isnull().sum()
print(f"Missing ICD7 Codes: {missing_icd_cd7}")
missing_icd_cd9 = icd_dgns_cd9.isnull().sum()
print(f"Missing ICD9 Codes: {missing_icd_cd9}")
missing_icd_cd11 = icd_dgns_cd11.isnull().sum()
print(f"Missing ICD11 Codes: {missing_icd_cd11}")
```

```python
data['ICD_PRCDR_CD25'].fillna('MISSING' , inplace=True)
data['HCPCS_CD'].fillna('MISSING' , inplace=True)
data['ICD_DGNS_CD1'].fillna('MISSING' , inplace=True)
data['ICD_DGNS_CD2'].fillna('MISSING' , inplace=True)
data['ICD_PRCDR_CD10'].fillna('MISSING' , inplace=True)
data['ICD_PRCDR_CD7'].fillna('MISSING' , inplace=True)
data['ICD_PRCDR_CD9'].fillna('MISSING' , inplace=True)
data['ICD_PRCDR_CD11'].fillna('MISSING' , inplace=True)
```

```python
print("Top 5 Most Common ICD Codes:\n", icd_frequency_cd25.head())
print("Top 5 Most Common ICD Codes:\n", hcpcs_frequency_cd.head())
print("Top 5 Most Common ICD Codes:\n", icd_frequency_cd1.head())
print("Top 5 Most Common ICD Codes:\n", icd_frequency_cd2.head())
print("Top 5 Most Common ICD Codes:\n", icd_frequency_cd10.head())
print("Top 5 Most Common ICD Codes:\n", icd_frequency_cd7.head())
print("Top 5 Most Common ICD Codes:\n", icd_frequency_cd9.head())
print("Top 5 Most Common ICD Codes:\n", icd_frequency_cd11.head())
```
